## Trasnform back to original variables

The variables with `Log` have been obtained originally by taking the `np.log10` of the variables. 

```python
tf_columns = ["Flux1.100m","Energy_Flux100","Significance","nu","nufnu","Variability_Index","Pivot_Energy"]
```

```python
new_data = 10**data_select[[f"Log{name}" for name in tf_columns]]
```

```python
log_names = [f"Log{name}" for name in tf_columns]
```

```python
dictionary = dict(zip(log_names, tf_columns))
dictionary
```

```python
new_data_renamed = new_data.rename(columns=dictionary)
```

```python
for col in ["X.name","z","Gaia_G_Magnitude","PL_Index","LP_Index","LP_beta","Label"]:
    new_data_renamed[col] = data_select[col]
```

```python
new_data_renamed
```

```python
new_data_renamed.to_csv("FullDataOriginal.csv")
```
